Const values:

```python
POPULATION_SIZE=250
P_CROSSOVER=0.9
P_MUTATION=0.0
MAX_GENERATIONS=100
```

Початкова популяція

```python
all=[i for i in range (0, len(D))]
copy_all=all.copy()
population=[copy_all]
population=np.array(population)
for i in range (1,POPULATION_SIZE):
  random.shuffle(copy_all)
  population=np.row_stack((population,copy_all))

print(np.array(D).shape)
print(np.array(population).shape)
print(population)
```

Пристосованість

```python
Fitness=np.zeros(POPULATION_SIZE)
for i in range (0,len(population)):
  for j in range (0,len(population[i])-1):
    Fitness[i]+=D[population[i][j]][population[i][j+1]]

print(Fitness)
print(Fitness.shape)
start_population = np.empty((0,26))
for i in range (0,len(population)):
  p=np.append(population[i],Fitness[i])
  #print("P: ",i,p[0])
  start_population=np.row_stack((start_population,p))

print(start_population)
```

```python
def calc_fitness(hrom):
  fit=0
  for j in range (0,len(hrom)-1):
    fit+=D[int(hrom[j])][int(hrom[j+1])]
  hrom.append(fit)
  return hrom
```

Вибір батьків:

*("Можна все на світі вибирати, сину, вибрати не можна тільки Батьківщину." (В. Симоненко))*




```python
def Tournament(population):
    inter_mas = []
    for n in range(POPULATION_SIZE):
        i1 = i2 = i3 = 0
        while i1 == i2 or i1 == i3 or i2 == i3:
            i1, i2, i3 = random.randint(0, POPULATION_SIZE-1), random.randint(0, POPULATION_SIZE-1), random.randint(0, POPULATION_SIZE-1)

        inter_mas.append(min([population[i1], population[i2], population[i3]], key=lambda x: x[len(D)]))

    return inter_mas
```

Схрещування

```python
def Crossing(par1, par2):
  p=random.randint(3, len(D[0])-2)
  ch1=[]
  for i in range(0,p):
    ch1.append(par1[i])
  for j in range(p,len(D[0])):
    if not(par2[j] in ch1):
      ch1.append(par2[j])
  if(len(ch1)<len(D[0])):
    for j in range(p,len(D[0])):
      if(not(par1[j] in ch1)):
        ch1.append(par1[j])
  ch1=calc_fitness(ch1)

  ch2=[]
  for i in range(0,p):
    ch2.append(par2[i])
  for j in range(p,len(D[0])):
    if not(par1[j] in ch2):
      ch2.append(par1[j])
  if(len(ch2)<len(D[0])):
    for j in range(p,len(D[0])):
      if(not(par2[j] in ch2)):
        ch2.append(par2[j])
  ch2=calc_fitness(ch2)

  return ch1, ch2
```

Mutation

```python
def Mutation(hrom):
  mp=random.randint(0,100)
  if(mp<P_MUTATION*100):
    rg1=random.randint(0,len(hrom)-2)
    rg2=random.randint(0,len(hrom)-2)
    hrom[rg1], hrom[rg2]=hrom[rg2], hrom[rg1]
  return hrom
```

Головна функція:

```python
generationCounter = 0
bestFitnessValues = []
while generationCounter < MAX_GENERATIONS:
    generationCounter += 1
    offspring = Tournament(start_population)
    start_population = np.empty((0, 26))

    for p1, p2 in zip(offspring[::2], offspring[1::2]):
        ch1, ch2=p1,p2
        if random.random() < P_CROSSOVER:
            ch1, ch2 = Crossing(p1, p2)
        if random.random() < P_MUTATION:
            ch1 = Mutation(ch1)

        if random.random() < P_MUTATION:
            ch2 = Mutation(ch2)
        #start_population = UpdatePopulation(start_population, ch1, ch2)
        start_population = np.row_stack((start_population, ch1))
        start_population = np.row_stack((start_population, ch2))

    fitnessValues = [ind[len(D)] for ind in start_population]

    bestFitness = min(fitnessValues)

    bestFitnessValues.append(bestFitness)
    print(f"Population N {generationCounter}: Path length = {bestFitness}")

    best_index = fitnessValues.index(min(fitnessValues))
    print("Best route = ", *start_population[best_index], "\n")
```

Елітний підхід відбору

Суть этого принципа заключается в том, что в новое поколение включаются лучшие родительские особи. Их число может быть от 1 и больше.

Использование "элитизма" позволяет не потерять хорошее промежуточное решение, но, в то же время, из-за этого алгоритм может "застрять" в локальном экстремуме

```python
sizeElite=20 #FUN FACT - зі 100 елітними особами у наступному поколінні - результат гірший ніж з 20

generationCounter = 0
bestFitnessValues = []
population=start_population
while generationCounter < MAX_GENERATIONS:
  generationCounter += 1  
  s=sorted(population,key=lambda x: x[len(D)])
  new_popul=np.empty((0,26))
  for i in range(0,sizeElite):
    new_popul=np.row_stack((new_popul,s[i]))
  
  while not len(new_popul)==POPULATION_SIZE:
        i1 = i2 = i3 = 0
        while i1 == i2:
            i1, i2 = random.randint(0, POPULATION_SIZE-1), random.randint(0, POPULATION_SIZE-1)
        c1, c2= Crossing(population[i1], population[i2])
        if random.random() < P_MUTATION:
            c1 = Mutation(c1)

        if random.random() < P_MUTATION:
            c2 = Mutation(c2)
        new_popul=np.row_stack((new_popul,c1))
        new_popul=np.row_stack((new_popul,c2))

  population=new_popul
  fitnessValues = [ind[len(D)] for ind in population]
  
  bestFitness = min(fitnessValues)

  bestFitnessValues.append(bestFitness)
  print(f"Population N {generationCounter}: Path length = {bestFitness}")

  best_index = fitnessValues.index(min(fitnessValues))
  print("Best route = ", *population[best_index], "\n")
```
